## 📌 What is an Activation Function?

An **activation function** is a mathematical function applied to the output of each neuron in a neural network. It introduces **non-linearity**, allowing the network to learn complex patterns that a purely linear model cannot capture.

Without activation functions, a deep neural network would simply be a linear transformation — no matter how many layers it has.

### ✅ Why do we need them?
- Introduce **non-linearity** into the model
- Allow learning of **complex mappings**
- Control the **output range** of neurons
- Help with **gradient flow** during backpropagation

## 1️⃣ Sigmoid Function

### Formula:
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

### Derivative:
$$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$$

### Key Properties:
- **Output range**: (0, 1) — good for probability outputs
- **Shape**: S-shaped (sigmoidal) curve
- **Saturates** at both ends → causes **vanishing gradient** problem
- **Not zero-centered** → slow convergence in deep networks
- Used in **binary classification** output layers

### ⚠️ Problems:
- **Vanishing gradients**: Gradients ≈ 0 for large |x|, killing backpropagation
- **Computationally expensive** due to exponential
- Outputs not zero-centered → zigzag gradient updates

````python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Sigmoid Activation Function', fontsize=16, fontweight='bold', color='#58a6ff', y=1.02)

# Function plot
axes[0].plot(x, sigmoid(x), color='#58a6ff', linewidth=3, label='σ(x)')
axes[0].axhline(0.5, color='#f78166', linestyle='--', alpha=0.7, label='y = 0.5')
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(1, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].fill_between(x, sigmoid(x), 0.5, alpha=0.15, color='#58a6ff')
axes[0].set_title('σ(x) = 1 / (1 + e⁻ˣ)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('σ(x)')
axes[0].legend(loc='upper left')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-0.1, 1.1)
axes[0].annotate('Saturates → \nVanishing Gradient', xy=(5, sigmoid(5)), xytext=(3, 0.85),
                  arrowprops=dict(arrowstyle='->', color='#f78166'), color='#f78166', fontsize=9)

# Derivative plot
axes[1].plot(x, sigmoid_derivative(x), color='#3fb950', linewidth=3, label="σ'(x)")
axes[1].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[1].fill_between(x, sigmoid_derivative(x), 0, alpha=0.2, color='#3fb950')
axes[1].set_title("Derivative: σ'(x) = σ(x)·(1−σ(x))", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("σ'(x)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
max_val = sigmoid_derivative(0)
axes[1].annotate(f'Max = {max_val:.3f}\nat x=0', xy=(0, max_val), xytext=(1.5, 0.22),
                  arrowprops=dict(arrowstyle='->', color='#f78166'), color='#f78166', fontsize=9)

plt.tight_layout()
plt.savefig('sigmoid.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print(f'\n📊 Sigmoid Statistics:')
print(f'  σ(-6)  = {sigmoid(-6):.6f}  (near 0)')
print(f'  σ(0)   = {sigmoid(0):.6f}  (exactly 0.5)')
print(f'  σ(6)   = {sigmoid(6):.6f}  (near 1)')
print(f'  Max gradient = {sigmoid_derivative(0):.4f} (at x=0)')
````

## 2️⃣ Tanh (Hyperbolic Tangent) Function

### Formula:
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} = 2\sigma(2x) - 1$$

### Derivative:
$$\tanh'(x) = 1 - \tanh^2(x)$$

### Key Properties:
- **Output range**: (-1, 1) — **zero-centered** ✅
- Stronger gradients than Sigmoid (max = 1.0 vs 0.25)
- Still suffers from **vanishing gradients** at extremes
- Used in **RNNs and LSTMs** as gate activations
- Generally **preferred over Sigmoid** for hidden layers

````python
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Tanh Activation Function', fontsize=16, fontweight='bold', color='#d2a8ff', y=1.02)

axes[0].plot(x, tanh(x), color='#d2a8ff', linewidth=3, label='tanh(x)')
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(1, color='#8b949e', linestyle=':', alpha=0.3, label='y = ±1')
axes[0].axhline(-1, color='#8b949e', linestyle=':', alpha=0.3)
axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].fill_between(x, tanh(x), 0, alpha=0.15, color='#d2a8ff')
axes[0].set_title('tanh(x) = (eˣ − e⁻ˣ) / (eˣ + e⁻ˣ)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('tanh(x)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Zero-centered ✓', xy=(0, 0), xytext=(1.5, -0.4),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=10, fontweight='bold')

axes[1].plot(x, tanh_derivative(x), color='#ffa657', linewidth=3, label="tanh'(x)")
axes[1].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[1].fill_between(x, tanh_derivative(x), 0, alpha=0.2, color='#ffa657')
axes[1].set_title("Derivative: 1 − tanh²(x)", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("tanh'(x)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].annotate(f'Max = 1.0 at x=0\n(4× stronger than Sigmoid)', xy=(0, 1), xytext=(1.5, 0.8),
                  arrowprops=dict(arrowstyle='->', color='#f78166'), color='#f78166', fontsize=9)

plt.tight_layout()
plt.savefig('tanh.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print(f'📊 Tanh Statistics:')
print(f'  tanh(-6)  = {tanh(-6):.6f}')
print(f'  tanh(0)   = {tanh(0):.6f}  (zero-centered)')
print(f'  tanh(6)   = {tanh(6):.6f}')
print(f'  Max gradient = {tanh_derivative(0):.4f} (at x=0)')
````

## 3️⃣ ReLU (Rectified Linear Unit)

### Formula:
$$\text{ReLU}(x) = \max(0, x)$$

### Derivative:
$$\text{ReLU}'(x) = \begin{cases} 1 & x > 0 \\ 0 & x \leq 0 \end{cases}$$

### Key Properties:
- **Output range**: [0, ∞)
- **Computationally very efficient** (just a threshold operation)
- Solves vanishing gradient for **positive** inputs
- Induces **sparsity** — some neurons output exactly 0
- **Default choice** for hidden layers in CNNs and feedforward networks

### ⚠️ Problems:
- **Dying ReLU**: Neurons that output 0 may never recover (gradient = 0)
- **Not zero-centered**
- **Unbounded** above — can cause exploding activations

````python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1.0, 0.0)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('ReLU Activation Function', fontsize=16, fontweight='bold', color='#3fb950', y=1.02)

axes[0].plot(x, relu(x), color='#3fb950', linewidth=3, label='ReLU(x)')
axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].fill_between(x, relu(x), 0, alpha=0.15, color='#3fb950')
axes[0].set_title('ReLU(x) = max(0, x)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('ReLU(x)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
# Shade dying ReLU region
axes[0].fill_between(x[x<=0], relu(x[x<=0]), 0, alpha=0.3, color='#f78166', label='Dead Zone')
axes[0].annotate('"Dead" neurons\n(gradient = 0)', xy=(-3, 0), xytext=(-5.5, 2),
                  arrowprops=dict(arrowstyle='->', color='#f78166'), color='#f78166', fontsize=9)
axes[0].legend()

axes[1].step(x, relu_derivative(x), color='#ffa657', linewidth=3, label="ReLU'(x)", where='post')
axes[1].fill_between(x, relu_derivative(x), 0, alpha=0.2, color='#ffa657', step='post')
axes[1].set_title("Derivative: 1 if x>0, else 0", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("ReLU'(x)")
axes[1].set_ylim(-0.2, 1.5)
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].annotate('Constant gradient (no vanishing!)', xy=(3, 1), xytext=(0.5, 1.3),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=9)

plt.tight_layout()
plt.savefig('relu.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print(f'📊 ReLU Statistics:')
print(f'  ReLU(-6) = {relu(-6):.1f}  (dead zone)')
print(f'  ReLU(0)  = {relu(0):.1f}   (kink point)')
print(f'  ReLU(6)  = {relu(6):.1f}   (linear)')
print(f'  Gradient x<0 = 0.0  |  Gradient x>0 = 1.0')
````

## 4️⃣ Leaky ReLU

### Formula:
$$\text{LeakyReLU}(x) = \begin{cases} x & x > 0 \\ \alpha x & x \leq 0 \end{cases}$$

where $\alpha$ is a small constant (typically 0.01)

### Derivative:
$$\text{LeakyReLU}'(x) = \begin{cases} 1 & x > 0 \\ \alpha & x \leq 0 \end{cases}$$

### Key Properties:
- **Fixes the Dying ReLU** problem by allowing small negative gradients
- **Output range**: (-∞, ∞)
- Still efficient to compute
- $\alpha$ is a **hyperparameter** (PReLU learns it during training)
- Better than ReLU when neurons keep dying during training

````python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1.0, alpha)

alphas = [0.01, 0.1, 0.3]
colors = ['#58a6ff', '#3fb950', '#ffa657']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Leaky ReLU Activation Function', fontsize=16, fontweight='bold', color='#ffa657', y=1.02)

for alpha, color in zip(alphas, colors):
    axes[0].plot(x, leaky_relu(x, alpha), color=color, linewidth=2.5, label=f'α = {alpha}')
    axes[1].step(x, leaky_relu_derivative(x, alpha), color=color, linewidth=2.5, label=f'α = {alpha}', where='post')

axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].set_title('LeakyReLU(x) = x if x>0, else αx', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('LeakyReLU(x)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Small slope (α) →\nNo dying neurons!', xy=(-3, leaky_relu(-3, 0.1)), xytext=(-5.5, -2),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=9)

axes[1].set_title("Derivative: 1 if x>0, else α", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("LeakyReLU'(x)")
axes[1].set_ylim(-0.1, 1.4)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('leaky_relu.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print('📊 Leaky ReLU (α=0.01) Statistics:')
print(f'  x = -6: {leaky_relu(-6):.4f}')
print(f'  x =  0: {leaky_relu(0):.4f}')
print(f'  x =  6: {leaky_relu(6):.4f}')
print(f'  Gradient x<0 = α = 0.01  |  Gradient x>0 = 1.0')
````

## 5️⃣ ELU (Exponential Linear Unit)

### Formula:
$$\text{ELU}(x) = \begin{cases} x & x > 0 \\ \alpha(e^x - 1) & x \leq 0 \end{cases}$$

### Derivative:
$$\text{ELU}'(x) = \begin{cases} 1 & x > 0 \\ \text{ELU}(x) + \alpha & x \leq 0 \end{cases}$$

### Key Properties:
- **Smooth** negative curve (no hard kink like Leaky ReLU)
- Output range: $(-\alpha, ∞)$
- **Negative saturation** helps noise robustness
- Converges faster than ReLU in practice
- More **computationally expensive** (exponential for x ≤ 0)

````python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1.0, elu(x, alpha) + alpha)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('ELU Activation Function', fontsize=16, fontweight='bold', color='#f78166', y=1.02)

alphas = [0.5, 1.0, 2.0]
colors = ['#58a6ff', '#f78166', '#d2a8ff']

for alpha, color in zip(alphas, colors):
    axes[0].plot(x, elu(x, alpha), color=color, linewidth=2.5, label=f'α = {alpha}')
    axes[1].plot(x, elu_derivative(x, alpha), color=color, linewidth=2.5, label=f'α = {alpha}')

axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].set_title('ELU(x) = x if x>0, else α(eˣ−1)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('ELU(x)')
axes[0].set_ylim(-3, 6)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Smooth negative saturation\n→ noise robust', xy=(-3, elu(-3)), xytext=(-5.5, -2.5),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=9)

axes[1].set_title("ELU Derivative", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("ELU'(x)")
axes[1].set_ylim(-0.5, 2.5)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('elu.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print('📊 ELU (α=1.0) Statistics:')
for val in [-6, -3, -1, 0, 3, 6]:
    print(f'  ELU({val:3d}) = {elu(val):.4f}')
````

## 6️⃣ Softmax Function

### Formula:
$$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{K} e^{x_j}}$$

### Key Properties:
- Converts a **vector of scores** into a **probability distribution**
- Output values are in (0, 1) and **always sum to 1** ✅
- Used exclusively in the **output layer** for multi-class classification
- Paired with **Cross-Entropy Loss**
- Amplifies differences: higher logits → much higher probability

### ⚠️ Note:
Softmax is applied to a **vector**, not element-wise like other activations.

````python
def softmax(x):
    e_x = np.exp(x - np.max(x))  # numerical stability
    return e_x / e_x.sum()

# Demonstrate with example logits
logits_examples = [
    np.array([1.0, 2.0, 3.0]),
    np.array([1.0, 4.0, 3.0]),
    np.array([1.0, 8.0, 3.0]),
]
classes = ['Cat', 'Dog', 'Bird']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Softmax Function — Multi-class Output Layer', fontsize=16, fontweight='bold', color='#79c0ff', y=1.02)

bar_colors = ['#58a6ff', '#3fb950', '#ffa657']

for i, (logits, ax) in enumerate(zip(logits_examples, axes)):
    probs = softmax(logits)
    bars = ax.bar(classes, probs, color=bar_colors, edgecolor='#30363d', linewidth=1.5)
    ax.set_ylim(0, 1.1)
    ax.set_title(f'Logits: {logits.tolist()}', fontweight='bold')
    ax.set_ylabel('Probability')
    ax.grid(True, alpha=0.3, axis='y')
    # Add value labels on bars
    for bar, prob in zip(bars, probs):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.02,
                f'{prob:.3f}', ha='center', va='bottom', fontsize=11, fontweight='bold', color='#e6edf3')
    ax.text(0.98, 0.02, f'Sum = {probs.sum():.4f}', transform=ax.transAxes,
            ha='right', va='bottom', color='#3fb950', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('softmax.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

# Temperature scaling demo
print('\n📊 Temperature Effect on Softmax:')
logits = np.array([2.0, 1.0, 0.1])
for T in [0.5, 1.0, 2.0, 5.0]:
    probs = softmax(logits / T)
    print(f'  T={T:.1f}: {[f"{p:.3f}" for p in probs]}  (sum={probs.sum():.3f})')
````

## 7️⃣ Swish Function

### Formula:
$$\text{Swish}(x) = x \cdot \sigma(\beta x) = \frac{x}{1 + e^{-\beta x}}$$

When $\beta = 1$: $\text{Swish}(x) = x \cdot \sigma(x)$

### Derivative:
$$\text{Swish}'(x) = \sigma(x) + x \cdot \sigma(x)(1 - \sigma(x))$$

### Key Properties:
- Discovered by **Google Brain** via neural architecture search (2017)
- **Smooth and non-monotonic** — unique among common activations
- Outperforms ReLU on deep networks in many tasks
- Slight **dip below zero** for small negative x
- Used in **EfficientNet, MobileNetV3**

````python
def swish(x, beta=1.0):
    return x * sigmoid(beta * x)

def swish_derivative(x, beta=1.0):
    s = sigmoid(beta * x)
    return s + x * beta * s * (1 - s)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Swish Activation Function', fontsize=16, fontweight='bold', color='#ffa657', y=1.02)

betas = [0.5, 1.0, 2.0]
colors = ['#58a6ff', '#ffa657', '#f78166']

for beta, color in zip(betas, colors):
    axes[0].plot(x, swish(x, beta), color=color, linewidth=2.5, label=f'β = {beta}')
    axes[1].plot(x, swish_derivative(x, beta), color=color, linewidth=2.5, label=f'β = {beta}')

# Also compare with ReLU
axes[0].plot(x, relu(x), color='#8b949e', linewidth=1.5, linestyle='--', label='ReLU (reference)', alpha=0.7)

axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].set_title('Swish(x) = x · σ(βx)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Swish(x)')
axes[0].set_ylim(-1, 6)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Non-monotonic dip\n(unique property)', xy=(-1.5, swish(-1.5)), xytext=(-5, -0.7),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=9)

axes[1].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[1].set_title("Swish Derivative", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("Swish'(x)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('swish.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print('📊 Swish (β=1) Statistics:')
for val in [-6, -2, -1, 0, 1, 2, 6]:
    print(f'  Swish({val:3d}) = {swish(val):.4f}')
min_x = x[np.argmin(swish(x))]
print(f'\n  Minimum at x ≈ {min_x:.2f}, value ≈ {swish(min_x):.4f}')
````

## 8️⃣ GELU (Gaussian Error Linear Unit)

### Formula:
$$\text{GELU}(x) = x \cdot \Phi(x) = x \cdot \frac{1}{2}\left[1 + \text{erf}\left(\frac{x}{\sqrt{2}}\right)\right]$$

Approximation:
$$\text{GELU}(x) \approx 0.5x\left(1 + \tanh\left[\sqrt{\frac{2}{\pi}}\left(x + 0.044715x^3\right)\right]\right)$$

### Key Properties:
- Used in **BERT, GPT, GPT-2, GPT-3, ViT** and most modern Transformers
- **Stochastic** interpretation: randomly zero-gates inputs
- Smoother than ReLU with a similar shape
- **State-of-the-art** for NLP and vision transformers
- Non-monotonic like Swish

````python
from scipy import special

def gelu(x):
    return x * 0.5 * (1.0 + special.erf(x / np.sqrt(2)))

def gelu_approx(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def gelu_derivative(x):
    return 0.5 * (1 + special.erf(x / np.sqrt(2))) + x * np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('GELU Activation Function (Used in Transformers)', fontsize=16, fontweight='bold', color='#79c0ff', y=1.02)

axes[0].plot(x, gelu(x), color='#79c0ff', linewidth=3, label='GELU (exact)')
axes[0].plot(x, gelu_approx(x), color='#ffa657', linewidth=2, linestyle='--', label='GELU (approx)', alpha=0.9)
axes[0].plot(x, relu(x), color='#8b949e', linewidth=1.5, linestyle=':', label='ReLU (ref)', alpha=0.7)
axes[0].axvline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[0].fill_between(x, gelu(x), 0, alpha=0.15, color='#79c0ff')
axes[0].set_title('GELU(x) = x · Φ(x)', fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('GELU(x)')
axes[0].set_ylim(-1, 6)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Smooth, like ReLU\nbut no hard zero-out', xy=(2, gelu(2)), xytext=(0.5, 4.5),
                  arrowprops=dict(arrowstyle='->', color='#3fb950'), color='#3fb950', fontsize=9)

axes[1].plot(x, gelu_derivative(x), color='#d2a8ff', linewidth=3, label="GELU'(x)")
axes[1].axhline(0, color='#8b949e', linestyle=':', alpha=0.5)
axes[1].fill_between(x, gelu_derivative(x), 0, alpha=0.2, color='#d2a8ff')
axes[1].set_title("GELU Derivative", fontweight='bold')
axes[1].set_xlabel('x')
axes[1].set_ylabel("GELU'(x)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('gelu.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()

print('📊 GELU vs GELU Approx Error:')
max_err = np.max(np.abs(gelu(x) - gelu_approx(x)))
print(f'  Max absolute error between exact and approx: {max_err:.6f}')
print(f'  (Very close — approximation used in practice for speed)')
````

## 🎯 How to Choose an Activation Function?

```
┌─────────────────────────────────────────────────────────────┐
│                  ACTIVATION FUNCTION DECISION GUIDE          │
├─────────────────────────────────────────────────────────────┤
│  Output Layer:                                              │
│    • Binary classification   → Sigmoid                      │
│    • Multi-class             → Softmax                      │
│    • Regression              → Linear (no activation)       │
│                                                             │
│  Hidden Layers:                                             │
│    • General / CNNs          → ReLU (fast, default)         │
│    • Dying neurons issue     → Leaky ReLU or ELU            │
│    • Transformers / NLP      → GELU                         │
│    • Deep networks           → Swish                        │
│    • RNNs / LSTMs            → Tanh + Sigmoid               │
│    • Avoid                   → Sigmoid / Tanh in deep nets  │
└─────────────────────────────────────────────────────────────┘
```

## ✅ Key Takeaways

| Problem | Solution |
|---------|----------|
| Vanishing gradient | Use ReLU, ELU, Swish, or GELU |
| Dying ReLU neurons | Use Leaky ReLU, ELU |
| Need probability output | Use Sigmoid (binary) or Softmax (multi-class) |
| Training transformers | Use GELU |
| Sparsity needed | Use ReLU |
| RNN memory gates | Use Tanh + Sigmoid |
